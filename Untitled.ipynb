```python
import os, glob
import cv2  # OpenCV for reading images
import tensorflow as tf  # TensorFlow
from tensorflow.keras import layers, models, optimizers, losses, metrics
import matplotlib.pyplot as plt
from collections import Counter
import seaborn as sns  # Seaborn for visualization
```

```python
IMAGE_SIZE= 256
BATCH_SIZE= 32
CHANNELS= 3
EPOCHS=20
```

```python
dataset=tf.keras.preprocessing.image_dataset_from_directory(
    "PLD_3_Classes_256/Training",
    shuffle= True,
    image_size=(IMAGE_SIZE,IMAGE_SIZE),
    batch_size=(BATCH_SIZE)
)
```

```python
class_names =dataset.class_names
print(class_names)
labels_list = []
for _, labels in dataset:
    labels_list.extend(labels.numpy())  # Convert tensor labels to a list

# Map labels to class names
original_labels = [class_names[label] for label in labels_list]

# Count occurrences of each class
class_counts = Counter(original_labels)

# Print formatted dataset-like structure
print("\nDataset Class Distribution:\n")
print(f"{'Class':<20}{'Count':<20}")
print("-" * 25)

for class_name, count in class_counts.items():
    print(f"{class_name:<20}{count:<20}")
```

```python
classes = list(class_counts.keys())
counts = list(class_counts.values())

# Plot the class distribution
plt.figure(figsize=(6, 4))
sns.barplot(x=classes, y=counts, palette="viridis")

# Add labels and title
plt.xlabel("Class", fontsize=14)
plt.ylabel("Number of Images", fontsize=14)
plt.title("Dataset Class Distribution", fontsize=16)
plt.xticks(rotation=45)  # Rotate x-axis labels if needed
plt.grid(axis='y', linestyle="--", alpha=0.7)

# Show the plot
plt.show()
```

```python
resizing_and_rescaling = tf.keras.Sequential([
    layers.Resizing(256, 256),          # Resize image to 256x256
    layers.Rescaling(1.0/255),          # Normalize pixel values to [0, 1]
])
```

```python
data_augmentation = tf.keras.Sequential([
    layers.RandomFlip("horizontal_and_vertical"),         # Random horizontal flip
    layers.RandomRotation(0.2),              # Random rotation (20%)
    # layers.RandomZoom(0.2),                  # Random zoom (20%)
    # layers.RandomTranslation(0.1, 0.1),      # Random shift in height and width (10%)
    # layers.RandomContrast(0.1)               # Slight contrast adjustment
])
```

```python
input_shape=(BATCH_SIZE ,IMAGE_SIZE, IMAGE_SIZE, CHANNELS)
n_classes=3
model =models.Sequential([
    resizing_and_rescaling,
    data_augmentation,
    layers.Conv2D(32,(3,3),activation='relu',input_shape=input_shape),
    layers.MaxPooling2D((2,2)),
    layers.Flatten(),
    layers.Dense(64,activation='relu'),
    layers.Dense(n_classes,activation='softmax'),
])
model.build(input_shape=input_shape)
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
history = model.fit(
    train_ds,
    epochs=EPOCHS,
    batch_size=BATCH_SIZE,
    verbose=1,
    validation_data=val_ds   
)
```

```python
scores = model.evaluate(test_ds)
```
